### Lineplot

```python
import matplotlib.pyplot as plt
input_integers = [1,2,3,4,5]
squares = [1,4,9,16,25]
plt.title("Plotting squares of first five positive integers", fontsize="12")
plt.xlabel("positive_integers", fontsize="8")
plt.ylabel("squares_of_integers", fontsize="8")
plt.tick_params(axis="both",which="major",labelsize="8")
plt.plot(squares, input_integers, linewidth="1", color="brown")
plt.show()
```

### Scatterplot

```python
import matplotlib.pyplot as plt
input_integers = [1,2,3,4,5]
squares = [1,4,9,16,25]
# Set marker size
marker_size = 25
plt.scatter(input_integers,squares,s=marker_size)
# set the chart title and axes
plt.title("Plotting squares of first five positive integers", fontsize="12")
plt.xlabel("positive_integers", fontsize="8")
plt.ylabel("squares_of_integers", fontsize="8")
plt.tick_params(axis="both",which="major",labelsize="8")
plt.show()
```

### Using Colormap

```python
import matplotlib.pyplot as plt
x_values = list(range(1001)) 
y_values = [x**2 for x in x_values] # list comprehension
def new_func(x_values, y_values):
    plt.scatter(x_values, y_values, s = 40, edgecolor = "none", \
             c=y_values, cmap = plt.cm.Blues)

new_func(x_values, y_values)
plt.savefig('squares',bbox_inches='tight')
```
